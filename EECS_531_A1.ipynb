1.1 (5 pts) Using the discrete definition, show that convolution is a linear operation, i.e.

$$
(f + g) * c = f*c + g*c \quad \textrm{and} \quad (\lambda f)*c = \lambda(f*c)
$$

<font color='Blue'>Answer part 1:</font>

We know from definition: 

$$output = o(i,j) = f(i,j)*h(k,l)=\sum_{k,l} f_{(i+k,j+l)}h_{(k,l)}$$

In here we simply replace $f(i,j)$ with $f(i,j)+g(i,j)$. Therefore we will have the expression: 

$$o(i,j) = (f(i,j)+g(i,j))*h(k,l)=\sum_{k,l} (f+g)_{(i+k,j+l)}h_{(k,l)}$$

We can use the rules of summation and get:

$$o(i,j) = \sum_{k,l} f_{(i+k,j+l)}h_{(k,l)}+ \sum_{k,l} g_{(i+k,j+l)}h_{(k,l)}$$

Which this is by definition:

$$output = o(i,j) = f(i,j)*h(k,l)+g(i,j)*h(k,l)$$

Hence proven:

$$(f + g) * c = f*c + g*c$$


<font color='Blue'>Answer part 2:</font>

We know that by definition:

$$output = o(i,j) = f(i,j)*h(k,l)=\sum_{k,l} f_{(i+k,j+l)}h_{(k,l)}$$

In here we simply replace $f(i,j)$ with $\lambda \times f(i,j)$. Therefore we will have the expression: 

$$o(i,j) = (\lambda \times f(i,j))*h(k,l)=\sum_{k,l} \lambda \times f_{(i+k,j+l)}h_{(k,l)}$$

We can use the rule of summation to extract a constant term:

$$o(i,j) = \sum_{k,l} \lambda \times f_{(i+k,j+l)}h_{(k,l)}=\lambda \times \sum_{k,l} f_{(i+k,j+l)}h_{(k,l)}$$

Therefore per definition of convolution under discrete time, we will have:

$$(\lambda f)*c = \lambda(f*c)$$

1.3. (5 pts) Describe the advantages and limitations of such an approach.

<font color='Blue'>Answer:</font>

The advantage approaching a 2D convolution is that it reduces the operation complexity. Previously, if we operate direcly as a 2D convolution, the operation complexity is $\mathbf{K^2}$. However, when we approach the problem as a one dimensional convolution followed by one dimensional convolution, the operation complexity is $\mathbf{2K}$. We can save the operation by one magnitude. (pg.102)

For an image with size $M \times N$ and a kernel size of $m \times n$ using traditional method will need a total of $M \times N \times m \times n$ multiplications and additions while for a seperatable kernal it only requires on the order of $M \times N \times (m+n)$ multiplications and additions. (book2 pg. 186)

The computational advantage is:

$$
C = \frac{mn}{m+n}
$$

However, in order to seperate a into a horizontal kernal and a vertical kernal requires prior knowledge that a kernal is seperatable. To prove that we will need to perform singular value decomposition on the kernal and if the singular value $\mathbf{\sigma_0}$ is non zero, then we can proceed to calculating the seperated vertical and horizontal kernal. (pg.103)
